```python
import pandas as pd
import dropbox
from dropbox.files import WriteMode
from datetime import datetime, timedelta
```

```python
ACCESS_TOKEN = 'your_dropbox_access_token'
```

```python
input_file_path = '/input_file/input_file.xlsx'
output_file_path = '/output_file/output_unique.xlsx'
```

```python
def process_file():
    dbx = dropbox.Dropbox(ACCESS_TOKEN)

    metadata, res = dbx.files_download(input_file_path)
    with open('input_file.xlsx', 'wb') as f:
        f.write(res.content)

    df = pd.read_excel('input_file.xlsx', sheet_name=0)

    df_unique = df.drop_duplicates(subset=['ORDERNO'], keep='first')

    df_unique.to_excel('output_unique.xlsx', index=False, engine='openpyxl')

    with open('output_unique.xlsx', 'rb') as f:
        dbx.files_upload(f.read(), output_file_path, mode=WriteMode('overwrite'))
```

```python
import schedule
import time

def job():
    print(f"Running the script at {datetime.now()}...")
    process_file()
    print("Script completed.")

# Karachi is UTC+5
schedule.every().day.at("03:00").do(job)  # 8 AM Karachi time is 3 AM UTC

while True:
    schedule.run_pending()
    time.sleep(1)
```
